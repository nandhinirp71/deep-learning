```python
import math
```

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(1)
```

```python
sigmoid(0)
```

```python
sigmoid(-1)
```

```python
sigmoid(-100)
# the below output means that it is very very close to
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(-100)
```

```python
tanh(0)
```

```python
tanh(100)
```

```python
tanh(-0.5)
```

```python
tanh(0.5)
```

```python
def relu(x):
  return max(0,x)
```

```python
relu(-100)
```

```python
relu(-0.5)
```

```python
relu(0)
```

```python
relu(0.5)
```

```python
relu(100)
```

```python
def leaky_relu(x):
  return max(0.1 * x, x)
```

```python
leaky_relu(-100)
```

```python
leaky_relu(-0.5)
```

```python
leaky_relu(0)
```

```python
leaky_relu(-0.5)
```

```python
leaky_relu(100)
```
