```python
# numpy supports multidimensional array operations
import numpy as np
```

```python
revenue = np.array([[100,200,220], [24,36,49], [12,18,20]])
expenses = np.array([[40, 90, 100], [10, 16, 20], [8, 10, 10]])
```

```python
profit = revenue - expenses
profit
```

```python
price_per_unit = np.array([1000, 400, 1200])
units = np.array([[30, 40, 50], [5, 10, 15], [2, 5, 7]])
```

```python
# numpy broadcasting
price_per_unit * units
```

```python
# numpy's matrix dot product
np.dot(price_per_unit, units)
```
