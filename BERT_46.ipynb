```python
import tensorflow_hub as hub
import tensorflow_text as text
```

```python
preprocess_url = "https://tfhub.dev/tensorflow/bert_en_uncased_preprocess/3"
encode_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/4"
```

```python
# creating function pointer
bert_preprocess_model = hub.KerasLayer(preprocess_url)
```

```python
text_test = ['nice movie indeed', "I love Python programming"]


text_preprocessed = bert_preprocess_model(text_test)
text_preprocessed.keys()
```

```python
text_preprocessed['input_mask']
# the output shape is (2, 128) because the input has 2 sentences
# and 128 because the max len of the sentence set by the model is 128

# And the length of 1st sentence is 3 words, where as the mask has 5 1s in the
# beginning because BERT includes
# CLS nice movie indeed SEP as prefix and suffix
```

```python
text_preprocessed['input_type_ids']
# BERT explicitly uses input_type_ids in its attention mechanism
# to control how different parts of the input interact with each other.
```

```python
text_preprocessed['input_word_ids']

# generates unique ids for eacg word
# CLS nice movie indeed SEP
```

After preprocessing we need to create another layer which has encoder url

```python
# creating function pointer
bert_model = hub.KerasLayer(encode_url)

# bert results is a dict
bert_results = bert_model(text_preprocessed)
bert_results.keys()
```

```python
bert_results['pooled_output']

# gives embedding vector for each of the 2 sentences in the input
# represents the meaning full sentense
# the output shape is 2*768 because 768 size of embedding vector
```

```python
bert_results['sequence_output']

# givs the embedding vector for each word
# the output shape is 2*120*768 because 128 is the max length of each sentence
# 768 is the length of embedding vector of each word
```

```python
bert_results['encoder_outputs'][0]
```

```python
len(bert_results['encoder_outputs'])

# length is 12 because we are using BERT base with 12 layers of Encoders
# each element in this represents the output of each encoder layer
```

```python
bert_results['encoder_outputs'][-1] == bert_results['sequence_output']

# last layer of encoder outputs is same as sequence output
```
