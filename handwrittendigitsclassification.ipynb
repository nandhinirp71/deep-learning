```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Importing handwritten dataset from Keras


```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

flatten the 2D array of images into 1D

```python
X_train.shape
# the output tells that X_train has 60,000 images that are stored as 28*28 pixels
```

```python
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train_flatten = X_train.reshape(len(X_train), 28*28)
```

```python
X_train_flatten.shape
```

```python
X_test_flatten = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flatten[0]
```

```python
# sequential means having a stack of neural network
# since it is stack so every layer of neural network would be
# an element in the stack
# in the below code 10 is the shape of output i.e., there are 10 categorial
# classes
# from 0 to 9
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=["accuracy"]
)

model.fit(X_train_flatten, y_train, epochs=9)
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_predicted = model.predict(X_test_flatten)
y_predicted[0]
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='.0%')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=False)
plt.xlabel('predicted')
plt.ylabel('Truth')
```

Using flatten feature by keras

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation="relu"),
    keras.layers.Dense(10,activation="sigmoid")
])

model.compile(
    optimizer='adam',
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.fit(X_train, y_train, epochs=5)
```
